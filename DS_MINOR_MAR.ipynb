```python
# pre processing

import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize


DataDir = '/content/images'
Categories = ['Football','Sunflower','Sunglasses']
target = []
flat_data = []
images = []

for i in Categories:
  class_num = Categories.index(i)
  path = os.path.join(DataDir,i)

  #print(path)  

  for img in os.listdir(path):
   # print(img)
   # to print the complete path of images we write -
   img_array = imread(os.path.join(path,img))
   # plt.imshow(img_array)
   # plt.show()
   img_resize = resize(img_array,(150,150,3))
   #plt.imshow(img_resize)
   #plt.show()
   flat_data.append(img_resize.flatten())
   images.append(img_resize)
   target.append(class_num)   
   
  
flat_data = np.array(flat_data)
target = np.array(target)  
images = np.array(images)
```

```python
target
```

```python
df = pd.DataFrame(flat_data)
df['Target'] = target
df.head()
```

```python
# Splitting data into Training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flat_data,target,test_size = 0.3,random_state = 0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000],'kernel':['linear']},
              {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']}
          ]

svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_pred,y_test)
```

```python
# saving the model
import pickle
pickle.dump(clf,open('Minor_Project.p','wb'))
```
